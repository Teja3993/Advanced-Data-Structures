```python
# Initial dictionary with 10 key-value pairs
student = {
    "id": 2025124988,
    "name": "Teja Karri",
    "age": 24,
    "department": "Engineering",
    "branch": "Data Science",
    "email": "karritej2@gmail.com",
    "phone": "77021044",
    "fees": 95000,
    "full_time": True,
    "location": "GITAM Visakhapatnam"
}

#Insertion and Update
def insert_update_pair(my_dict,key,value):
  print("Length before adding new pair ",len(my_dict))
  print("Inserting key ",key," with value ",value)
  my_dict[key] = value
  print("Length after adding new pair ",len(my_dict))
  return my_dict

#Looking up inside dictionary
def get_value(my_dict,key):
  print("Getting value of ",key)
  return my_dict[key]

#Looking up inside dictionary using get()
def get_value_g(my_dict,key):
  print("Looking up ",key)
  return my_dict.get(key,"Not found")

#Deleting key value pairs
def del_value(my_dict,key):
  print("Deleting key ",key," with value ",my_dict[key])
  del my_dict[key]
  return my_dict

#Deleting complete dictionary
def delete_dict(my_dict):
  print("Deleting all contents in dictionary")
  my_dict.clear()
  return my_dict

#Deleting using pop()
def delete_pop(my_dict,key):
  print("Deleting key ",key)
  return my_dict.pop(key)

#Checking if key is present in dictionary
def is_member(my_dict,key):
  print("Checking if key ",key," is present in dictionary ")
  return key in my_dict

#Implementing iteration in dictionary

def print_values(my_dict):
  print("List of all values in given dictionary")   #printing values
  for val in my_dict.values():
    print(val)

def print_keys(my_dict):
  print("List of all keys in given dictionary")          #printing keys
  for k in my_dict.keys():
    print(k)

def print_keys_values(my_dict):
  print("List of all key-value pairs in given dictionary")    #printing key-value pairs
  for k,v in my_dict.items():
    print(k,":",v)


print(insert_update_pair(student,"Roll No",33))



print(get_value(student,"phone"))

print(get_value_g(student,"address"))

print(del_value(student,"full_time"))

print(delete_pop(student,"age"))

print(is_member(student,"parent_name"))

print_values(student)
print_keys(student)
print_keys_values(student)


print(delete_dict(student))





```
